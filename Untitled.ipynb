```python
from bs4 import BeautifulSoup
import requests
```

```python
request = requests.get("https://www.olx.com.br/brasil")
```

```python
soup = BeautifulSoup(request.text, 'html.parser')
```

```python
category_navbar = soup.find(class_="search-category-nav")
```

```python
category_navbar_links = category_navbar.findAll('a')
```

```python
category = {}
for category_ in category_navbar_links:
    category[category_.get('title')] = category_.get('href').split('/')[-1] #pega somente a extensao para acessar o site
#     category[category_.get('title')] = [category_.get('href'), category_.get('href').split('/')[-1]]    
```

```python
category_extension = list(category.values())
```

```python
category_extension
```

```python
links_table = soup.find(class_='linkshelf-tabs-content country')
```

```python
states_links = links_table.findAll('a')
```

```python
states = {}
for link in states_links:
    states[link.text] = link.get('href')
```

```python
for category_ in category_extension:
    for state_ in states.values():
        print(state_+'/'+category_)
```
